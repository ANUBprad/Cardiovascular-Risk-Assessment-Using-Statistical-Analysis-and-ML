```python
import pandas as pd
from sklearn.impute import SimpleImputer

df = pd.read_csv(r"C:\Users\ANUBHAV\OneDrive\Apps\Cardiovascular_Risk_Assessment_Using_Statistical_Analysis_and_ML\data\raw\cardio_vascular.csv")
```

```python
# Handle missing values
num_cols = df.select_dtypes(include=["int64","float64"]).columns
imputer = SimpleImputer(strategy="median")
df[num_cols] = imputer.fit_transform(df[num_cols])
```

```python
# Feature engineering
df["age_group"] = pd.cut(
    df["age"], [0,40,55,100],
    labels=["Young","Middle","Senior"]
)

df["age_chol_interaction"] = df["age"] * df["totChol"]
```

```python
# Encode categorical variables
df = pd.get_dummies(df, drop_first=True)
```

```python
# Save clean dataset
df.to_csv(r"C:\Users\ANUBHAV\OneDrive\Apps\Cardiovascular_Risk_Assessment_Using_Statistical_Analysis_and_ML\data\processed\cleaned_featured_data.csv", index=False)
```
