```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\ANUBHAV\OneDrive\Apps\Cardiovascular_Risk_Assessment_Using_Statistical_Analysis_and_ML\data\raw\cardio_vascular.csv ")
```

```python
df.shape
df.info()
df.describe()
```

```python
#Age v Disease

plt.figure(figsize=(6,4))
sns.boxplot(x="TenYearCHD", y="age", data=df)
plt.title("Age vs Cardiovascular Disease")
plt.xlabel("Disease (0 = No, 1 = Yes)")
plt.ylabel("Age")
plt.show()
```

```python
#Cholesterol vs Disease

plt.figure(figsize=(6,4))
sns.boxplot(x="TenYearCHD", y="totChol", data=df)
plt.title("Cholesterol vs Cardiovascular Disease")
plt.show()
```

```python
#Heart Rate vs Disease

plt.figure(figsize=(6,4))
sns.boxplot(x="TenYearCHD", y="heartRate", data=df)
plt.title("Heart Rate vs Cardiovascular Disease")
plt.show()
```

```python
#Correlation Heatmap

plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm", annot=False)
plt.title("Feature Correlation Heatmap")
plt.show()
```
