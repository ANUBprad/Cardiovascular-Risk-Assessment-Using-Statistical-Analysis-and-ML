```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu

df = pd.read_csv(r"C:\Users\ANUBHAV\OneDrive\Apps\Cardiovascular_Risk_Assessment_Using_Statistical_Analysis_and_ML\data\raw\cardio_vascular.csv")
```

```python
# Age hypothesis test
ttest_ind(
    df[df["TenYearCHD"] == 1]["age"],
    df[df["TenYearCHD"] == 0]["age"]
)
```

```python
# Cholesterol hypothesis test (non-parametric)
mannwhitneyu(
    df[df["TenYearCHD"] == 1]["totChol"],
    df[df["TenYearCHD"] == 0]["totChol"]
)
```

```python
# Heart rate hypothesis test
ttest_ind(
    df[df["TenYearCHD"] == 1]["heartRate"],
    df[df["TenYearCHD"] == 0]["heartRate"]
)
```
